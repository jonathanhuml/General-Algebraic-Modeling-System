```python
%%gams

set i /cc, bf/;
set t /m1,m2,m3/;

parameters
hold(i) /cc 0.5, bf 0.4/;

table c(i,t)  "supply cost"
              m1       m2      m3   
    cc        3        3.4     3.8      
    bf        2.5      2.8     3.4;
    
table d(i,t)  "demand cost"
              m1       m2      m3   
    cc        40       30      20      
    bf        20       30      10;


free variables cost "total cost";

positive variables x(i,t), l(i,t) "leftover cakes";

equations
obj, demand(i,t), supply(t), remainder(i,t);

obj..cost =e= sum((i,t), c(i,t)*(x(i,t)+l(i,t))) + sum(t, sum(i, hold(i)*l(i,t)));

demand(i,t).. x(i,t)+l(i,t-1)$(not sameas(t,'l')) =g= d(i,t);

supply(t).. sum(i, x(i,t)+l(i,t)) =l= 65; 

remainder(i,t).. l(i,t) =e= x(i,t)-d(i,t);

model cakes /all/;

solve cakes using lp minimizing cost;
```
